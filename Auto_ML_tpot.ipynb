```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tpot import TPOTRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error
from sklearn.preprocessing import PolynomialFeatures
# Load your dataset
data = pd.read_csv('/content/drive/MyDrive/Data set/Prasanna raj KJ_717822I239.csv')
data=data.replace(np.nan,0)
# Separate features and target variable
X = data.iloc[:,[0,2,3,5]] # Features
y = data.iloc[:,[1]] # Target variable
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
# Define polynomial features
poly = PolynomialFeatures(degree=6)  # You can adjust the degree of polynomial
# Transform features to polynomial features
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
tpot = TPOTRegressor(generations=5, population_size=20, verbosity=2, random_state=1) # You can adjust generations and population_size
# Train the TPOT model
tpot.fit(X_train_poly, y_train)
# Make predictions
predictions = tpot.predict(X_test_poly)
# Evaluate the model
print("R2 score:", r2_score(y_test, predictions))
# Calculate evaluation metrics
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mse)
msle = mean_squared_log_error(y_test, predictions)
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Squared Logarithmic Error (MSLE):", msle)
```
