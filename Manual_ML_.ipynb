```python
x=data.iloc[:,[0,2,3,5]]
y=data.iloc[:,[1]]
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
a=LinearRegression()
a.fit(xtrain,ytrain)
```

```python
ypred=a.predict(xtrain)
```

```python
from sklearn.metrics import r2_score,mean_squared_error as mse,mean_absolute_error, mean_squared_log_error
print("R2_score",r2_score(ytrain,ypred))
print("The mean square Error:",mse(ytrain,ypred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
p=PolynomialFeatures(degree=6)
x_poly=p.fit_transform(x)
a.fit(x_poly,y)
poly_pred=a.predict(x_poly)
```

```python
print("R2 score of the Polynomial regression with degree 6 is:", r2_score(y,poly_pred))
# Calculate evaluation metrics
mse = mse(y,poly_pred)
mae = mean_absolute_error(y,poly_pred)
rmse = np.sqrt(mse)
msle = mean_squared_log_error(y,poly_pred)
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Squared Logarithmic Error (MSLE):", msle)
```
